```python
import pandas as pd
import numpy as np
from decimal import *

order1 = pd.read_csv('F:/泰迪/outfile.csv', encoding='gbk')
order1['支付时间'] = pd.to_datetime(order1['支付时间'])
order1 = pd.DataFrame(order1)
#先把饮料类分开并提取出来
order1 = order1.groupby('大类')  
order2 = order1.get_group('饮料')
order3 = order1.get_group('非饮料')

print('按大类分组后的订单详情表为：', order2.head(),order3.head())  
```

```python
order4 = order2[['商品','设备ID','订单数量']].groupby('设备ID')
one = order4.get_group('E43A6E078A04172')  #A
two = order4.get_group('E43A6E078A04134')  #B
three = order4.get_group('E43A6E078A04228') #C
four = order4.get_group('E43A6E078A07631') #D
five = order4.get_group('E43A6E078A06874') #E
```

```python
def fun(x):
    if x >= 200:
        return ('热销')
    
    if 100 <= x < 200:
        return ('正常')
        
    else:
        return ('滞销')



```

```python
one = one.groupby('商品')
one = one.size().reset_index(name='订单数量')
one = one.sort_values(by = '订单数量',ascending=False)
a = np.zeros((112,1))
for i in range(0,112):
    a[i] = i+1

one.insert(0,'序号',a.astype(int))
one['订单数量'] = one['订单数量'].apply(lambda x: fun(x)).rename('标签')
print('机器A饮料类商品的标签：', one)
```
